```python
import numpy as np
```

```python
import pandas as pd
```

```python
from sklearn.model_selection import train_test_split
```

```python
#loading the dataset into the environment, header='None' since there are no headers in the dataset
sonar=pd.read_csv('/content/sonar_data (1).csv',header=None)
```

```python
sonar.head()
```

```python
sonar.shape
```

```python
#gives you the count of each class label i.e. M and R
sonar[60].value_counts()
```

```python
sonar.groupby(60).mean()
```

```python
#seperate dependent data
Y=sonar[60]
```

```python
Y
```

```python
#copying independent data into X
X=sonar.drop(columns=60,axis=1)
```

```python
X.shape
```

```python
#split into train and test split,
#stratify will split data so that there are equal number of mines and rocks in the training data
#random state is to split the data in a specific order/manner
X_train,X_test,Y_train,Y_test=train_test_split(X,Y,test_size=0.1,stratify=Y,random_state=2)
```

```python
X_train.shape
```

```python
X_test.shape
```

```python
Y_train.shape
```

```python
Y_test.shape
```

```python
#import random forest classifier
from sklearn.ensemble import RandomForestClassifier
```

```python
#estimator represents the number of random trees we use
model=RandomForestClassifier(n_estimators=100)
```

```python
#train the model using the training data
model.fit(X_train,Y_train)
```

```python
#prediction on X_test which is not seen by the model
Y_pred_unseen=model.predict(X_test)
```

```python
Y_pred_unseen
```

```python
#predicting on X_train , this data is already seen by the model
Y_pred_seen=model.predict(X_train)
```

```python
Y_pred_seen
```

```python
#Testing the accuracy
from sklearn.metrics import accuracy_score
```

```python
print(" Accuracy on seen data: ",accuracy_score(Y_train,Y_pred_seen)*100,"%")
```

```python
print(" Accuracy on unseen data: ",accuracy_score(Y_test,Y_pred_unseen)*100,"%")
```
